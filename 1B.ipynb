```python

 def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and arr[low] <= key <= arr[high]:
        probes += 1

        # Avoid division by zero
        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            return -1, probes

        # Estimate the probable position
        pos = low + int(
            ((high - low) * (key - arr[low])) /
            (arr[high] - arr[low])
        )

        if arr[pos] == key:
            return pos, probes

        if arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes

        if arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


# Main Program
n = int(input("Enter number of student roll numbers: "))

print("Enter sorted roll numbers:")
roll_numbers = list(map(int, input().split()))

if len(roll_numbers) != n:
    print(f"Error: Please enter exactly {n} roll numbers.")
    exit()

key = int(input("Enter roll number to search: "))

interp_index, interp_probes = interpolation_search(roll_numbers, key)
binary_index, binary_probes = binary_search(roll_numbers, key)

print("\nInterpolation Search:")
if interp_index != -1:
    print(f"Roll number found at index {interp_index}")
else:
    print("Roll number not found")
print(f"Probes required = {interp_probes}")

print("\nBinary Search:")
if binary_index != -1:
    print(f"Roll number found at index {binary_index}")
else:
    print("Roll number not found")
print(f"Probes required = {binary_probes}")

print("\nComparison:")
if interp_probes < binary_probes:
    print("Interpolation Search used fewer probes.")
elif interp_probes > binary_probes:
    print("Binary Search used fewer probes.")
else:
    print("Both searches used the same number of probes.")
```
